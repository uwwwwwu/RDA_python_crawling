```python
from selenium import webdriver
from bs4 import BeautifulSoup
import calendar
import datetime, time
import pandas
from openpyxl import Workbook

#data = []
num = 0
driver = webdriver.Chrome('C:\Chrome-driver\chromedriver.exe')
write_wb = Workbook()
write_ws = write_wb.active
write_ws['A1'] = '날짜'

dt_index = pandas.date_range(start='20100101', end='20190530')
dt_list = dt_index.strftime("%Y-%m-%d").tolist()

for date in dt_list:
    data = []
    print(date)
    page = 0
    i = 1
    while i != 0:
        page = page + 1
        driver.get('https://www.kahis.go.kr/home/lkntscrinfo/selectLkntsOccrrncList.do?pageIndex='+str(page)+'&turmGubun=02&occrFromDt='+str(date)+'&occrToDt='+str(date)+'&dissCl=&ctprvn=+&signgu=+&lstkspCl=&legalIctsdGradSe=')
            
        print("접속")
        
        div_elems = driver.find_elements_by_xpath("//*[@id='homeLkntscrinfoVO']/table[5]/tbody/tr[3]/td/table/tbody/tr/td")
        
        elems = driver.find_elements_by_xpath("//*[@id='homeLkntscrinfoVO']/table[4]/tbody/tr[2]/td/table/tbody/tr/td")
        for j in elems:
            if j.text == '검색내용이 존재하지 않습니다.':
                print("존재아냐")
                i = 0
        
        if i == 0 :
            break
        html = driver.page_source
        soup = BeautifulSoup(html,'html.parser')
        div = soup.find("div",{"class":"oldtable"})
        table = div.find_all("table")
        menu_table = table[9]
            
        for tr in menu_table.find_all('tr'):
            tds = list(tr.find_all('td'))
                
            for td in tds:
                a = tds[0].text
                b = tds[1].text
                b2 = b.strip()
                c = tds[2].text
                d = tds[3].text
                e = tds[4].text
                f = tds[5].text
                g = tds[6].text
            data.append([str(date),a,b2,c,d,e,f,g])
            write_ws.append([str(date),a,b2,c,d,e,f,g])

        i = i - 1
        for div_el in div_elems:
            print(div_el.text)
            join_text = ''.join(div_el.text)
            join_out = join_text.split('/')
        
            if '2쪽)' in join_out:
                print("2쪽")
                i = i + 1
                if page == 2:
                    i = 0
                
            if '3쪽)' in join_out:
                print("3쪽")
                i = i + 1
                if page == 3:
                    i = 0
                    
    for i in data:
        print(i)
write_wb.save('C:\\Users\\PC\\Desktop\\병역\\가축질병.xlsx')
```
